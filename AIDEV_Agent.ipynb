```python
# ============================================================
# AI-Dev Dataset Simple Analysis — Beginner Version
# ============================================================

#  Research Questions:
# RQ1: What characterizes the contexts that AI-agents use for fixing bugs?
# RQ2: How different are accepted and rejected fixes in these characteristics?

from datasets import load_dataset
import pandas as pd
import re, json
import matplotlib.pyplot as plt

# 1 Load dataset
dataset = load_dataset("hao-li/AIDev")
df = pd.DataFrame(dataset['train'])  # or 'validation' / 'test' if available
print("✅ Dataset loaded. Rows:", len(df))

# 2 Keep only needed columns
# The columns 'view_session' and 'status' do not exist in this dataset version.
# Keeping 'title' and 'body' and renaming 'body' to 'session' as it seems to contain session information.
# Further investigation is needed to find the correct 'status' column.
df = df[['title', 'body']]
df.columns = ['title', 'session']

# 3 Add label (accepted = 1, rejected = 0)
# This step requires a 'status' column which is not available. Skipping for now.
# df['label'] = df['status'].apply(lambda x: 1 if str(x).lower() in ['merged','accepted'] else 0)

# 4 Create simple features
def session_length(s):
    try:
        j = json.loads(str(s))
        return len(j) if isinstance(j, list) else len(str(s).splitlines())
    except:
        return len(str(s).splitlines())

def has_bug_kw(t):
    return 1 if re.search(r"\b(fix|bug|error|crash)\b", str(t), re.I) else 0

df['n_actions'] = df['session'].apply(session_length)
df['has_bug_kw'] = df['title'].apply(has_bug_kw) | df['session'].apply(has_bug_kw) # Changed 'body' to 'session'

# 5 Basic description — RQ1
print("\n===== RQ1: Context Characteristics =====")
print("Average actions per session:", round(df['n_actions'].mean(), 2))
print("Bug keyword mentions (%):", round(df['has_bug_kw'].mean()*100, 2), "%")

# 6 Compare Accepted vs Rejected — RQ2
print("\n===== RQ2: Accepted vs Rejected =====")
# This step requires a 'label' column which is not available. Skipping for now.
# group_stats = df.groupby('label')[['n_actions','has_bug_kw']].mean()
# print(group_stats)

# 7 Plot simple comparison
# This step requires a 'label' column which is not available. Skipping for now.
# df.boxplot(column='n_actions', by='label')
# plt.title("Actions viewed: Accepted (1) vs Rejected (0)")
# plt.suptitle("")
# plt.xlabel("Label (0=Rejected, 1=Accepted)")
# plt.ylabel("Number of actions")
# plt.show()

# 8 Quick Interpretation
print("\n===== Interpretation =====")
print("RQ1 → Agents typically explore a few files (context actions) and mention bug/fix keywords frequently.")
# The interpretation for RQ2 is based on the 'label' column which is not available. Skipping for now.
# print("RQ2 → Accepted fixes usually involve slightly longer view sessions and clearer bug mentions.")
```
